```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img1 = mpimg.imread("spw25.png")
img2 = mpimg.imread("spw27.png")
img3 = mpimg.imread("spw27(CN).png")
img4 = mpimg.imread("spw29.png")
img5 = mpimg.imread("spw31.png")
```

```python
fig, axs = plt.subplots(3, 2, figsize=(20, 20)) 
axs[0, 0].imshow(img1)
axs[0, 0].set_title('spw25')
axs[0, 0].axis('off')

axs[0, 1].axis('off')

axs[1, 0].imshow(img2)
axs[1, 0].set_title('spw27')
axs[1, 0].axis('off')

axs[1, 1].imshow(img3)
axs[1, 1].set_title('spw27(CN)')
axs[1, 1].axis('off')

axs[2, 0].imshow(img4)
axs[2, 0].set_title('spw29')
axs[2, 0].axis('off')

axs[2, 1].imshow(img5)
axs[2, 1].set_title('spw31')
axs[2, 1].axis('off')
```
